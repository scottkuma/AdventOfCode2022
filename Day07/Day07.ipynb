```python
class Directory:
    def __init__(self, name, parent):
        self.name = name
        self.size = None
        self.contents = {}
        self.parent = parent
        self.type = 'dir'
        
    def __repr__(self):
        return f"dir {self.name}"
    
    def add_subdir(self, name):
        new_subdir = Directory(name, self)
        self.contents[name] = new_subdir
        
        
    def add_file(self, name, size):
        new_file = File(name, size, self)
        self.contents[name] = new_file  
        
        
    def get_size(self):
        tot_size = 0
        for lbl,obj in self.contents.items():
            if obj.type == 'dir':
                tot_size += obj.get_size()
            elif obj.type == 'file':
                tot_size += obj.size
                
        #print(self.name, tot_size)
        return tot_size
            

        
class File:
    def __init__(self, name, size, parent):
        self.name = name
        self.size = int(size)
        self.parent = parent
        self.type = 'file'
        
    def __repr__(self):
        return f"{self.size} {self.name}"

# Global Vars
    
root = Directory("/", None)
pos = root    
    
    
# Commands

def cd(name):
    global pos
    global root
    if name in pos.contents and pos.contents[name].type=='dir':
        
        pos = pos.contents[name]
    elif name == ".." and pos.parent != None:
        
        pos = pos.parent
    elif name == "/":
        
        pos = root
    elif name not in pos.contents:
        print(f"ERROR: No such file or directory '{name}'.")
    elif pos.contents[name].type!='dir':
        print(f"ERROR: {name} is not a directory.")
    
def pwd():
    global pos
    print(pos.name)
    return None

def ls():
    global pos
    if pos.name != "/":
        print("..")
    if len(pos.contents.keys()) > 0:
        for k in pos.contents.keys():
            print(pos.contents[k])
    else:
        print("--EMPTY DIR--")
        
def mkdir(name):
    global pos
    pos.add_subdir(name)
    
def touch(name, size):
    global pos
    pos.add_file(name, size)

def sizeof():
    global pos
    #print(pos)
    #print(pos.contents)
    return pos.get_size()
        
def process_line(lp):
    global pos
    #print(lp)
    if lp[0] == "$":
        if lp[1] == "cd":
            cd(lp[2])
        elif lp[1] == "ls":
            ls()
    else:
        if lp[0].isnumeric():
            touch(lp[1], lp[0])
        elif lp[0] == 'dir':
            mkdir(lp[1])
            
```

```python
with open("data.txt") as infile:
    for line in infile:
        line = line.strip()
        print(line)
        line_parts = line.split(' ')
        process_line(line_parts)
```
