```python
import pandas as pd
import numpy as np
import io
```

```python
def process_header(header):
    header.reverse()
    out = ''.join(header)
    df = pd.read_fwf(io.StringIO(out))
    
    s = {}
    
    for c in df.columns:
        s[c] = df.loc[:,c].dropna()
    
    return s



def move_box_9000(stacks, moveFrom, moveTo):
    
    mf = stacks[str(moveFrom)].dropna()
    mt = stacks[str(moveTo)].dropna()
    
    box = mf[mf.index.max()]
    mf=mf.drop(index=mf.index.max())
    
    mt = pd.concat([mt,pd.Series(box)],ignore_index=True)
    
    return (mf, mt)

```

```python
with open("manifest.txt") as infile:
    header = []
    while True:
        line = infile.readline()
        if line == "\n":
            break
        else:
            header.append(line)

    stacks = process_header(header)
    
    #print(stacks)
    
    while True:
        line = infile.readline()
        
        if len(line) == 0:
            break
        
        parts = line.strip().split(' ')
        #print(parts)
        times = int(parts[1])
        mf = parts[3]
        mt = parts[5]
        
        for x in range(times):
            #print(x + 1)
            stacks[mf], stacks[mt] = move_box_9000(stacks, mf, mt)

out = ""
for x in stacks:
    xs = stacks[x][stacks[x].index.max()]
    out += xs[1]
print(out)

            
```

```python
def move_box_9001(stacks, moveFrom, moveTo, num):
    
    mf = stacks[str(moveFrom)].dropna()
    mt = stacks[str(moveTo)].dropna()
    
    boxes = mf[-num:]
    mf = mf[:-num]
    mt = pd.concat([mt,pd.Series(boxes)],ignore_index=True)

    return (mf, mt)


with open("manifest.txt") as infile:
    header = []
    while True:
        line = infile.readline()
        if line == "\n":
            break
        else:
            header.append(line)

    stacks = process_header(header)
    
    while True:
        line = infile.readline()
        
        if len(line) == 0:
            break
        
        parts = line.strip().split(' ')
        #print(parts)
        #print(stacks)
        times = int(parts[1])
        mf = parts[3]
        mt = parts[5]
        
        stacks[mf], stacks[mt] = move_box_9001(stacks, mf, mt, times)

out = ""
for x in stacks:
    xs = stacks[x][stacks[x].index.max()]
    out += xs[1]
print(out)
```
