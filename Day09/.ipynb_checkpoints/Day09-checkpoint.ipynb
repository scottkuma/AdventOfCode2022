```python
class RopePart:
    def __init__(self):
        self.pos_x = 0
        self.pos_y = 0
        pos_tuple = self.pos_x, self.pos_y
        self.visited = set()
        self.visited.add(pos_tuple)
    
    def get_pos(self):
        return (self.pos_x, self.pos_y)
    
    def move(self,moveTuple):
        (move_x, move_y) = moveTuple
        self.pos_x += move_x
        self.pos_y += move_y
        self.visited.add(self.get_pos())
    
    def __repr__(self):
        out = str( self.get_pos()) + " / " + str(self.visited)
        return out
    
class Rope:
    dx =  {"UL": (-1,1),
           "U": (0,1),
           "UR": (1,1),
           "R": (1,0),
           "DR": (1,-1),
           "D": (0,-1),
           "DL": (-1,-1),
           "L": (-1,0) 
          }
    
    def __init__(self):
        self.head = RopePart()
        self.tail = RopePart()
    
    def __repr__(self):
        return f"ROPE:\tHead @ {self.head}\n\t Tail @ {self.tail}"
    
    def move_tail(self):
        (hx,hy) = self.head.get_pos()
        (tx,ty) = self.tail.get_pos()
        r = ( (hx - tx) , (hy-ty) )
        #print("Res",r)
        if abs(r[0]) > 1 or abs(r[1]) > 1:
            if r == (2,0):
                self.tail.move(self.dx['R'])
            elif r == (-2,0):
                self.tail.move(self.dx['L'])
            elif r == (0,2):
                self.tail.move(self.dx['U'])
            elif r == (0,-2):
                self.tail.move(self.dx['D'])
            elif r in [(1,2), (2,1)]:
                self.tail.move(self.dx['UR'])
            elif r in [(2,-1), (1, -2)]:
                self.tail.move(self.dx['DR'])
            elif r in [(-1,-2), (-2,-1)]:
                self.tail.move(self.dx['DL'])
            elif r in [(-2,1), (-1,2)]:
                self.tail.move(self.dx['UL'])
            else:
                exit(1)
        else:
            pass
    
    def move_head(self, moveStr):
        self.head.move(self.dx[moveStr])
        self.move_tail()

```

```python
import pandas as pd
import numpy as np

rope = Rope()

df = pd.read_csv("data.txt",sep=' ',names=['dir','num'])

for idx, row in df.iterrows():
    for x in range(row.num):
        rope.move_head(row.dir)
        
print("The tail visited", len(rope.tail.visited), "positions")
```
