```python
class Monkey():
    verbose = False
    def __init__(self, barrel):
        self.name = len(barrel)
        self.barrel = barrel
        self.items = None
        self.op = None
        self.testFactor = None
        self.ifTrue = None
        self.ifFalse = None
        self.itemsExamined = 0
        self.superfactor = None
        
    def __repr__(self):
        out = f"Monkey{self.name}:\n"
        out += str(self.items) + " Test Factor: " + str(self.testFactor)
        #out += "\nOp: " + str(self.op)
        out += f"\nTrue: {self.ifTrue}  False: {self.ifFalse}"
        out += f"\n# of Monkeys in barrel: {len(self.barrel)}"
        out += "\n"
        return out
    
    def testItem(self, itemLevel):
        if itemLevel % self.testFactor == 0:
            return True
        else:
            return False
    
    def setOp(self, op):
        self.op = lambda old: eval(op)
    
    def catchItem(self, itemLevel):
        if self.verbose: 
            print(f"    Monkey {self.name} catches an item level {itemLevel}")
        self.items.append(itemLevel)
        
    def throwItem(self, itemLevel, throwto):
        if self.verbose: 
            print(f"    Monkey {self.name} throws item to Monkey {throwto}")
        
        #only throw the mod of the itemlevel to the superfactor
        self.barrel[throwto].catchItem(itemLevel % self.superfactor)
    
    def examineItem(self):
        itemLevel = int(self.items.pop(0))
        if self.verbose: 
            print(f"Monkey {self.name} Examines an item with worry level {itemLevel}")
        self.itemsExamined += 1
        
        itemLevel = self.op(itemLevel)
        if self.verbose: 
            print(f"    Worry Level to {itemLevel}!!"  if self.verbose else "")
    
        #itemLevel = itemLevel // 3
        if self.verbose: 
            print(f"    Monkey gets bored with item. It doesn't help: worry level remains {itemLevel}")
        
        if self.testItem(itemLevel):
            if self.verbose: 
                print(f"    Worry level IS divisible by {self.testFactor}")
            self.throwItem(itemLevel, int(self.ifTrue))
        else:
            if self.verbose: 
                print(f"    Worry level IS NOT divisible by {self.testFactor}")
            self.throwItem(itemLevel, int(self.ifFalse))
        
        
barrel = []
factors = []

filename = "data.txt"
with open(filename, 'r') as infile:
    for line in infile:
        line = line.strip().split(':')
        if "Monkey" in line[0]:
            currMonkey = Monkey(barrel)
            barrel.append(currMonkey)
        
        elif "Starting items" in line[0]:
            items = line[1].split(',')
            items = [int(x) for x in items]
            #print(items)
            currMonkey.items = items

        elif "Operation" in line[0]:
            fn = line[1].split('=')
            #print(fn[1])
            op = fn[1]
            barrel[-1].setOp(op)
            
        elif "Test" in line[0]:
            factor = line[1].split(' ')[-1]
            currMonkey.testFactor = int(factor)
            factors.append(int(factor))
        
        elif "If true" in line[0]:
            currMonkey.ifTrue = int(line[1][-1])
            
        elif "If false" in line[0]:
            currMonkey.ifFalse = int(line[1][-1])
            
#print(factors)
superfactor = 1
for f in factors:
    superfactor *= f

for m in barrel:
    m.superfactor = superfactor

for round in range(10000):
    if (round+1) % 1000 == 0:
        print(round+1)
    for x in barrel:
        #print(f"r{round} m{x.name}: {x.items}")
        for item in range(len(x.items)):
            x.examineItem()
    #print(f"Finished Round {round+1}")
    

monkey_items = []    
for x in barrel:
    monkey_items.append(x.itemsExamined)

print(monkey_items)
    
monkey_items.sort()

print(f"\n\nShenanigans Level = {monkey_items[-1] * monkey_items[-2]}")
```
